```python
from bs4 import BeautifulSoup 
html_sample = ' \                  # below is example content of a html file
<html> \
<body> \
<h1 id="title">Hello World</h1> \
<a href="#" class="link">This is link1</a> \
<a href="# link2" class="link">This is link2</a> \
</body> \
</html>'

soup = BeautifulSoup(html_sample, 'lxml')
print(soup.text)
```

```python
soup = BeautifulSoup(html_sample, 'lxml')
header = soup.select('h1')
print(header)
print(header[0])
print(header[0].text)
```

```python
soup = BeautifulSoup(html_sample, 'lxml')
alink = soup.select('a')
print(alink)
for link in alink:
    print(link)
    print(link.txt)
```

```python
alink = soup.select('#title')    # find link with id "title"
print(alink)
```

```python
soup = BeautifulSoup(html_sample, 'lxml')  # find class of 'link'
for link in soup.select('.link'):
    print(link)
```

```python
alinks = soup.select('a')
for link in alinks:
    print(link['href'])
```
