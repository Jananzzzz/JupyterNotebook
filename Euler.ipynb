```python
import matplotlib.pyplot as plt

x, y = 0.0, 1.0
step = 0.2
X, Y = [x], [y]

def f(x:float, y:float):
    if x == 0.0:
        return -1
    return -y - x * y**2

while x < 6.0:
    k = f(x, y)
    y = -k - (x+step) * y**2
    x = x + step
    X.append(x)
    Y.append(y)
    
plt.plot(X, Y)
plt.show()
```
